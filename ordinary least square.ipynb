```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('cpp_cleaned_data')
df.head()
```

```python
x=df.iloc[:,1:54]
y=df['price']
```

```python
x.iloc[0]-2
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
```

```python
num=0
den=0
for i in range(1,205):
    num = num + ((x.iloc[i]-x_mean)*(y[i]-y_mean))
    den+= (x.iloc[i]-x_mean)**2

m=num/den
c=y-m*x_mean
print(f"m: {m}\nc: {c}")    
```
